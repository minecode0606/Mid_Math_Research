# 구면의 방정식

spherical surface equation
구면(球面)의 방정식은 3차원 방정식의 일종으로 3차원 공간에서 한 점에 대하여
일정거리에 있는 점들의 집합이다.

다음은 구면의 방정식의 그래프중 일종이다.

## 상세

중점이 O(a,b,c)이고 반지름이 r인 구면의 방정식은

√{(x - a)^2 + (y - b)^2 + (z - c)^2} = r
로 나타낼수 있다.

### 유도과정

위 방정식은 3차원 공간에서 피타고라스의 정리를 활용하여 직육면체에서 대각선의 길이를 구하는 공식에서 유도된다.

##### 피타고라스 정리

직각삼각형에서 빗변의 길이를 c 빗변이 아닌 다른 두 변의 길이를 각각  a , b 라고 하면 다음이 성립한다.

a^2 + b^2 = c^2
이떄 양변에 제곱근을 씌우면

√(a^2 + b^2)= c가 성립한다.

objectbox2.jpeg

(변 FG) = x, (변 GH) = y, (변 HD) = z라 하자.
피타로라스 정리의 의하여
* 직각삼각형 FGH에서 (변 FH) = √(x^2 + y^2)
* 직각삼각형 FHD에서 (변 FD : 대각선의 길이) =  √(x^2 + y^2 + z^2)

따라서 직육면체에서의 대각선의 길이 : 3차원 공간에서 두 점 사이의 거리는
√(x^2 + y^2 + z^2)이라고 정의할수 있다.

##### 응용
이걸 방정식으로 옮기면 거리가 일정한 점들의 집합이라고 할수 있는데,

√(x^2 + y^2 + z^2) = r인 경우 3차원 입체공간에서 원점 O(0,0,0)에서
r만큼 떨어진 점들의 자취의 방정식이 된다.

### 구의 겉넓이

위 자취의 방정식이 그리는 점들의 집합의 넓이를 S라 하자, 그럼, S는  구의 겉넓이와 일치한
4πr^2이라 할수 있다.

이때 반지름은 음수일수 없으므로
r∈{r|r은 양수}일때
* S = 4πr^2
가 성립한다.

이를 그래프로 표현해보자,

```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.font_manager as fm
def add_font(fontsize = 15):
    ## 폰트 이름
    title_font_name = 'NanumGothic'
    tick_label_font_name = title_font_name

    ## 폰트 경로
    font_path = './NanumGothic.ttf'
    ## 폰트 사이즈
    font_size = fontsize


    ## FontProperties 인스턴스 생성
    font_prop = fm.FontProperties(fname = font_path, size=font_size)
    return font_prop

font_prop = add_font()
class Surfacearea():
    def __init__(self, radiusend = 10):
        self.perimeter = 3.141592
        self.radius = np.linspace(0,radiusend, 200)

    @classmethod
    def Finding_the_S_value(cls, radiusls, perimeter):
        outputls = []
        for radius in radiusls:
            outputls.append(4 * perimeter * (radius ** 2))
        return outputls

    def drow(self):
        S = self.Finding_the_S_value(self.radius, self.perimeter)
        plt.subplot(2,1,2)
        plt.xlabel("radius(반지름)", fontproperties=font_prop)
        plt.ylabel("area(구의 겉넓이)", fontproperties=font_prop)
        plt.plot(self.radius, S, label = "Graph of area of a sphere \nas a function of radius")
        plt.legend()
        plt.subplot(2,1,1)
        plt.title("반지름에 따른 구면의 넓이 그래프 (위 : 반지름 음수 포함) (아래 : 반지름 음수 제외)", fontproperties=font_prop)
        radius2 = np.linspace(-max(self.radius), max(self.radius), 200)
        S2 = self.Finding_the_S_value(radius2, self.perimeter)
        plt.plot(radius2, S2)
        plt.show()

surface = Surfacearea(10)
surface.drow()
```

위에 그래프를 보면 r∈{r|r은 실수 전체}일때 S = 4πr^2의 그래프는 r에 대해서 축의 방정식이 r = 0이고 아래로 볼록한 이차함수의 곡선을
나타내는것을 알수 있습니다.

하지만 위 가정에서 반지름은 음수가 될수 잆으므로 r∈{r|r은 양수}일때 축의 방정식 r = 0으로부터
오른쪽에 위치한 곡선인 아래 그래프가 해당됩니다.

```python
class Drowpoint(Surfacearea):
    def point(self):
        graphfont = add_font(10)
        S = self.Finding_the_S_value(self.radius, self.perimeter)
        plt.xlabel("radius(반지름)", fontproperties=font_prop)
        plt.ylabel("area(구의 겉넓이)", fontproperties=font_prop)
        plt.plot(self.radius, S)
        plt.annotate('상대적으로 기울기가 작음',xy=(self.radius[20], S[20]),xytext=(3,20),arrowprops={'color':'red'}, fontproperties=graphfont)
        plt.annotate('상대적으로  기울기가 큼',xy=(self.radius[180], S[180]),xytext=(4, 1100),arrowprops={'color':'red'}, fontproperties=graphfont)
        plt.show()

point = Drowpoint()
point.point()
```
